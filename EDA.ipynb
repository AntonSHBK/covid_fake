```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download stopwords if not already downloaded
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')


# Define stop words
stop_words = set(stopwords.words('english'))

# Function to remove URLs from text
def remove_links(text):
    if isinstance(text, str):  # Ensure text is a string
        return re.sub(r'.*(http|www).*', '', str(text))
    return ''  # Return original if not a string


# Function to remove stopwords
def remove_stopwords(text):
    if isinstance(text, str):
        words = word_tokenize(str(text))  # Tokenize text
        filtered_text = " ".join([word for word in words if word.lower() not in stop_words])
        return filtered_text
    return ''

# Apply function to the "Message" column
df_cleaned = pd.DataFrame()

df_cleaned['Message'] = df['Message'].apply(remove_stopwords)
df_cleaned['Description'] = df['Description'].apply(remove_stopwords)
df_cleaned['Message'] = df_cleaned['Message'].apply(remove_links)
df_cleaned['Description'] = df_cleaned['Description'].apply(remove_links)

df_cleaned
```

```python
arr = []
for i, text in enumerate(df['Message']):
    for j, description in enumerate(df['Description']):
        if text == description:
            if i not in arr:
                arr.append(i)
            if j not in arr:
                arr.append(j)
```
